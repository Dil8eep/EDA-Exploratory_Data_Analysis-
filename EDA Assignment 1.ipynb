```python
#7. How does the prevalence of high blood pressure differ between male and female patients?
```

```python
data.columns
```

```python
male = data[data["sex"]==1]
female = data[data["sex"]==0]
male_bp_count = male["high_blood_pressure"].sum()
female_bp_count = female["high_blood_pressure"].sum()
print("Prevalence of high blood pressure among male patients:",male_bp_count/len(male))
print("Prevalence of high blood pressure among female patients:",female_bp_count/len(female))
```

```python
sns.countplot(data=data,x='sex',hue='high_blood_pressure')
```

```python
# >> female have high prevalence of blood pressure
```
